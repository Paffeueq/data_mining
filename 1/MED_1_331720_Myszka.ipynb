> **Zadanie 1**  Znajdź:
* państwa o zbliżonej liczbie ludności do Polski
* najbardziej podobne do Polski państwa świata pod względem zarówno powierzchni jak i liczby ludności

```python
# Zadanie 1a
polska = panstwa[panstwa['Country'] == 'Poland'].iloc[0]
pop_polski = polska['Population 2025']

prog_min = pop_polski * 0.9
prog_max = pop_polski * 1.1

podobna_populacja = panstwa[
    (panstwa['Population 2025'] >= prog_min) & 
    (panstwa['Population 2025'] <= prog_max) &
    (panstwa['Country'] != 'Poland')
]

print("Panstwa o podobnej liczbie ludnosci do Polski:")
print(podobna_populacja[['Country', 'Population 2025']])
```

```python
# Zadanie 1b
panstwa['PopDiff'] = abs(panstwa['Population 2025'] - pop_polski)
panstwa['AreaDiff'] = abs(panstwa['Land Area (KmÂ²)'] - polska['Land Area (KmÂ²)'])

panstwa['SimilarityScore'] = panstwa['PopDiff'] + panstwa['AreaDiff']

najbardziej_podobne = panstwa[panstwa['Country'] != 'Poland'].sort_values('SimilarityScore').head(5)

print("\nNajbardziej podobne panstwa do Polski (powierzchnia + ludnosc):")
print(najbardziej_podobne[['Country', 'Population 2025', 'Land Area (KmÂ²)', 'SimilarityScore']])
```

> **Zadanie 2** Skonfrontuj, dla obu powyższych zbiorów, podstawowe miary danych z wykresami: histogramem, wykresem pudełkowym i punktowym (macierzy wykresów punktowych), dla kompletnych zbiorów i dla poszczególnych klas w tych zbiorach. Zlokalizuj poszczególne miary na wykresach. Oceń łatwość lokalizacji poszczególnych miar na każdym rodzaju wykresu. Jakie wnioski dotyczące każdego ze zbiorów danych można wyciągnąć z powyższych obserwacji ?

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# =====================
# ZBIOR D
kolumny_d = ['atrybut1', 'atrybut2']

# Histogramy
d[kolumny_d].hist(bins=10, figsize=(8,4))
plt.suptitle('Histogramy - D (caly zbior)')
plt.tight_layout()
plt.show()

# Boxplot
d[kolumny_d].plot(kind='box', figsize=(8,4), title='Boxplot - D (caly zbior)')
plt.tight_layout()
plt.show()

# Scatter matrix (kolory wg klas)
kolory_d = d['klasa'].map({'klasa 1':'red','klasa 2':'green','klasa 3':'blue'})
scatter_matrix(d[kolumny_d], figsize=(8,8), diagonal='hist', c=kolory_d)
plt.suptitle('Scatter matrix - D (caly zbior)')
plt.show()

# Dla poszczegolnych klas
for kl in d['klasa'].unique():
    d[d['klasa']==kl][kolumny_d].hist(bins=10, figsize=(8,4))
    plt.suptitle(f'Histogramy - D ({kl})')
    plt.tight_layout()
    plt.show()

    d[d['klasa']==kl][kolumny_d].plot(kind='box', figsize=(8,4), title=f'Boxplot - D ({kl})')
    plt.tight_layout()
    plt.show()

# =====================
# ZBIOR IRIS
kolumny_iris = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Histogramy
iris[kolumny_iris].hist(bins=10, figsize=(8,6))
plt.suptitle('Histogramy - Iris (caly zbior)')
plt.tight_layout()
plt.show()

# Boxplot
iris[kolumny_iris].plot(kind='box', figsize=(8,6), title='Boxplot - Iris (caly zbior)')
plt.tight_layout()
plt.show()

# Scatter matrix (kolory wg gatunku)
kolory_iris = iris['species'].map({s:i for i,s in enumerate(iris['species'].unique())})
scatter_matrix(iris[kolumny_iris], figsize=(8,8), diagonal='hist', c=kolory_iris)
plt.suptitle('Scatter matrix - Iris (caly zbior)')
plt.show()

# Dla poszczegolnych gatunkow
for gatunek in iris['species'].unique():
    iris[iris['species']==gatunek][kolumny_iris].hist(bins=10, figsize=(8,6))
    plt.suptitle(f'Histogramy - Iris ({gatunek})')
    plt.tight_layout()
    plt.show()

    iris[iris['species']==gatunek][kolumny_iris].plot(kind='box', figsize=(8,6), title=f'Boxplot - Iris ({gatunek})')
    plt.tight_layout()
    plt.show()
```

Histogramy dobrze pokazują rozkład wartości każdej cechy oraz wartości skrajne.


Wykresy pudełkowe pozwalają łatwo odczytać medianę, rozstęp oraz wartości odstające.


Macierze wykresów punktowych najlepiej pokazują zależności między cechami i różnice między klasami lub gatunkami.


Dla zbioru d oraz iris klasy/gatunki są wyraźnie oddzielone na scatter matrix, co pozwala zauważyć ich różnice.


Lokalizacja mediany i kwartylów jest najłatwiejsza na boxplotach, a zależności między cechami najlepiej widać na scatter matrix.

> **Zadanie 5** Skonfrontuj macierz korelacji z wykresami punktowymi zbiorów `d` oraz `iris`. Jaka jest zależność między konfiguracją punktów na wykresie punktowym a wartością miary korealcji ? Jakie wnioski odnoszące się do poszczególnych atrybutów można wyciągnąć na podstawie analizy macierzy korelacji ?  

```python


# =======================
# 1. Zbior d - wykresy punktowe
# =======================
# 'd' ma kolumny 'atrybut1', 'atrybut2', 'klasa'
sns.pairplot(d, hue='klasa', vars=['atrybut1', 'atrybut2'])
plt.suptitle('Wykresy punktowe zbioru d', y=1.02)
plt.show()

# Macierz korelacji dla zbioru d (tylko kolumny numeryczne)
d_numeric = d[['atrybut1', 'atrybut2']]
d_corr = d_numeric.corr()
print("Macierz korelacji dla zbioru d:")
print(d_corr)

# Heatmapa korelacji
plt.figure(figsize=(4,4))
sns.heatmap(d_corr, annot=True, cmap='coolwarm')
plt.title('Heatmapa korelacji zbioru d')
plt.show()

# =======================
# 2. Zbior iris - wykresy punktowe
# =======================
# wybieramy kolumny numeryczne
iris_numeric = iris.select_dtypes(include='number')
sns.pairplot(iris, hue='species', vars=iris_numeric.columns)
plt.suptitle('Wykresy punktowe zbioru iris', y=1.02)
plt.show()

# Macierz korelacji dla zbioru iris
iris_corr = iris_numeric.corr()
print("Macierz korelacji dla zbioru iris:")
print(iris_corr)

# Heatmapa korelacji
plt.figure(figsize=(6,6))
sns.heatmap(iris_corr, annot=True, cmap='coolwarm')
plt.title('Heatmapa korelacji zbioru iris')
plt.show()
```

Zaleznosc konfiguracja punktow a korelacja:

Silna korelacja (|ρ|≈1): punkty ukladaja sie w wyrazna, waska linie prosta, np. petal_length i petal_width w zbiorze iris (ρ≈0.96).

Brak korelacji liniowej (|ρ|≈0): punkty sa rozproszone bez wyraznego liniowego wzorca lub ukladaja sie w struktury nieliniowe.

W zbiorze d punkty tworza klastry, ale wzorzec nie jest liniowy, dlatego ρ≈−0.034. Korelacja nie wychwytuje silnej zaleznosci strukturalnej.

W zbiorze iris cechy sepal_length i sepal_width sa prawie liniowo niezalezne (ρ≈−0.12).

Wnioski z macierzy korelacji:

Zbior d: atrybuty sa liniowo niezalezne (ρ≈−0.034), co oznacza, ze zmiany jednej zmiennej nie sa zwiazane z druga w sposob liniowy. Jest to zgodne z rozproszonym ukladem punktow.

Zbior iris: wystepuje silna dodatnia korelacja (ρ≈0.82–0.96) miedzy wszystkimi wymiarami platka (petal_length, petal_width) i dlugoscia kielicha (sepal_length). Wymiary te przekazuja podobne informacje.
Najbardziej niezalezny jest sepal_width (ρ bliskie 0 z sepal_length, umiarkowanie ujemne z wymiarami platka, ρ≈−0.43), co oznacza, ze dostarcza unikalnej informacji.
Brak silnych korelacji ujemnych; najsilniejsza ujemna zaleznosc wystepuje miedzy sepal_width a petal_length (ρ≈−0.43).
